# Oficina Aula 7

Olá, seja bem-vindo!

Durante a aula sobre a Nuvem de Palavras, você conheceu esse tipo de gráfico e aprendeu a criá-lo com Kibana, além de utilizá-lo para visualizar muitas quantidades de categorias. Agora é hora de exercitar o que você aprendeu!

Em primeiro lugar, caso você ainda não tenha o Elasticsearch e o Kibana instalados na sua máquina, realize a instalação desses sistemas no seu computador antes de iniciar esta oficina. Isso será necessário para carregar os dados de amostra Sample flight data, disponibilizados pelo Kibana, em Sample data.

A partir desses dados, crie uma visualização do tipo Nuvem de Palavras (Tag cloud), na qual as tags representem os aeroportos de origem dos voos (Origin), e a métrica, associada ao tamanho de cada tag, seja o total de minutos de atraso nos voos (FlightDelayMin) por aeroportos.
