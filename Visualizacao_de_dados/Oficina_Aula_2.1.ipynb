```python
# Resposta da questão 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

```python
# Resposta da questão 2
contagens = filmes.groupby(by='Genre').count()
plt.figure(figsize=(15,5))
sns.barplot(x="Genre", y="Budget_M", color = 'darkviolet', ci = 'sd', data=filmes)
plt.title('Orçamento de todos os filmes')
plt.show()
```

```python
# Resposta da questão 3
filmes.sort_values(by='Worldwide_Gross_M', ascending = False).head()
plt.figure(figsize=(15,5))
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'darkred', ci = 'sd', data=filmes)
plt.title('A bilheteria de todos os filmes')
plt.show()
```

```python
# Resposta da questão 4
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', data=filmes)
plt.show()
```

```python
# Resposta da questão 5
plt.figure(figsize=(15,8))
plt.title('Média de orçamento por gênero em cada ano')
sns.barplot(x="Year", y = 'Budget_M' , ci = None, hue= 'Genre', data=filmes)
plt.show()
```
