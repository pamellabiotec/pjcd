```python
# Questão 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
df = sns.load_dataset('mpg')
df.head()
```

```python
# Questão 2 - Relação entre a aceleração e o consumo de combustível dos veículos com a diferenciação do país de origem
sns.relplot(x="acceleration", y="mpg", col = 'origin', data=df)
plt.show()

# Respostas: Sim existe uma correlação, embora apresenta no Japão sem. Ainda podemos classificar o EUA tem baixa correlação 
# positiva, Japão é sem correlação e a Europa tem alta correlação positiva.Logo, não são as mesmas para todos os países.
```

```python
# Questão 3
plt.figure(figsize=(20,7))
plt.title('Relação entre a aceleração e o consumo de combustível dos veículos')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=df)
plt.show()

# Resposta: Sim, existe uma relação entre as três variáveis, onde percebe-se que os modelos dos anos de 80 e 84 apresentam o 
# consumo mais alto mesmo com pouca aceleração. Já os modelos dos anos de 76, 72 e 68 apresentam em média um consumo baixo em
# em relação a aceleração. Embora há algumas exceções.
```
