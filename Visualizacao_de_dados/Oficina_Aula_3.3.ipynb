```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
print ("Setup completo")
```

```python
# Questão 1
df_filepath = "dataset/dadosoficina07.txt"
df = pd.read_csv(df_filepath, parse_dates=True)
df.tail()
```

```python
# Questão 2

plt.figure(figsize=(22,5))
plt.title('Série temporal das vendas ao longo dos meses')
sns.lineplot(x="Month", y="Sales", data=df)
plt.show()

# Houve uma tendência crescente que iniciou em 1-11, contendo uma flutuação até 3-12.
```
