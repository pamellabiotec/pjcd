```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
print("Setup Completo")
```

```python
# Questão 1
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Questão 2
plt.figure(figsize=(15,7))
plt.title('Distribuição do valor da passagem de acordo com a cidade de embarque')
sns.violinplot(x="embark_town", y="fare", data=titanic)
plt.show()
```

```python
# Questão 3 - Distribuição do valor da passagem, de acordo com a cidade de embarque, sexo e se o passageiro sobreviveu ou não.
plt.figure(figsize=(15,7))
sns.catplot(x="embark_town", y="fare", hue = 'sex', col = 'survived', kind = 'violin', split = True, data=titanic)
plt.show()

# Resposta: Podemos notar a diferença entre os sobreviventes e os náufragos, onde a maioria das mulheres e homens que  
# embarcaram em Cherbourg sobreviveram.E que é relativamente proporcional os sobreviventes e náugragos, homens e mulheres 
# que embarcaram em Southampton. 
```
