```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Questão 1
df = pd.read_csv('dadosoficina05.txt')
df.head()
```

```python
# Questão 2
plt.figure(figsize=(15, 5))
plt.title('Distribuição dos valores de aluguel')
sns.kdeplot(df['rent amount (R$)'], shade=True)
plt.show()
```

```python
# Questão 3
plt.figure(figsize=(10, 5))
plt.title('Distribuição dos valores de seguro contra incêndio')
sns.distplot(df['fire insurance (R$)'])
plt.show()
```
