### Introdução à Árvore de Decisão

Olá

Nesta aula, iremos começar a estudar Mais um modelo de aprendizado de máquina conhecido como a **Árvore de Decisão**, Decision Trees. 

O modelo Árvores de decisão e um dos mais interessantes devido a sua capacidade de interpretabilidade. É um modelo que possui uma ideia bem intuitiva e, também é altamente interpretável por um especialista.Nós podemos construir este modelo a partir de um conjunto de dados. Porém, uma vez consutrído este modelo conseguimos entender qual é a sua motivaçãopara retornar uma classe ou um valor no caso de uma regressão.

Então, as árvores de decisão podem ser bastante utilizadas juntamente com um especialista, que pode avaliar se as decisões que a árvore está retornando são, de fato, decisões que possuem algum embasamento real. Uma vez que a teoria e a prática podem divergir um pouco.

As Árvores de Decisão são conhecidas como conjunto de regras "se-então" utilizadas para regressão e classificação. Iremos exemplificar ao utilizar um conjunto de dados sobre peixe. Onde cada nó seria uma regra. 



Como obter a árvore a partir dos dados?

Utiliza-se um índice de impureza do resultados de uma divisão. Vários índices podem ser utilizados (gini, entropia...) No caso, iremos utilizar o **índice Gini**. É o índice bastante utilizado na literatura. Inclusive é o índice padrão da nossa biblioteca **Scikit-learn**.

Então o índice Gini funciona da seguinte maneira: ele é calculado por 1 menos o somatório da divisão dos nossos dados ao quadrado para cada classe.

## Implementando a Árvore de Decisão

Nessa aula iremos implementar a **Árvore de Decisão**, onde iremos aplicar a versão de classsificação quanto a versão de regressão.

Então, iremos iniciar importando as bibliotecas que iremos utilizar.


``
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
``

Dessa maneira, estamos importando a biblioteca de árvore de decisão tanto para classificação quanto para regressão na mesma linha. Em seguida, iremos importar os datasets que iremos utilizar. Sendo Boston para regressão e Iris para classificação.


``
from sklearn.datasets import load_boston, load_iris
``

Na sequência iremos aplicar a divisão de treino e teste.


``
from sklearn.model_selection import train_test_split
``

Para implementar a árvore de decisão para classificação utilizaremos o seguinte comando:


``
dt = DecisionTreeClassifier()
``

Em seguida, iremos dividir os conjunto de dados Iris em treino e teste usando a função train_test_split.

``
X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y)
``

Onde X receberá os atributos e o y receberá o rótulo dos dados. Em seguida, iremos aplicar a função de divisão de treino e teste.

Na sequência, implementaremos a árvore de decisão para classificação:

``
dt.fit(X_train,y_train)
``

Ao executar essa linha, foi treinada nossa árvore de decisão para classificação.  Agora, iremos utilizar a função **score** para verificar essa implementação.

``
dt.score(X_test,y_test)
``

Ao aplicar essa função, teremos como resultado 94% de precisão. Agora, iremos aplicar a Árvore de decisão para regressão utilizando passos bem parecidos. 


``
dt = DecisionTreeRegressor()
``

Assim, instanciamos nosso modelo para regressão.
Nesse momento, iremos carregar o dataset Boston com seguinte código:

``
X,y = load_boston(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X,y)

``

Na sequência implementaremos a árvore de decisão para regressão:


``
dt.fit(X_train,y_train)
``

E verificaremos nosso modelo com a função score:


``
dt.score(X_test,y_test)
``

Assim, teremos como resultado que nosso modelo teve 69% de acerto.

O primeiro ponto que precisamos verificar é que a árvore de decisão por natureza já possui um alto **índice de overfitting**, porque , pela sua natureza ela tenta sempre carregar todo o conjunto de dados etenta separá-lo perfeitamente. Então, teremos que aplicar alguma técnica para minizar o efeito de overfitting. 

Logo, a partir deste momento, irei retreinar a árvore de decisão utilizando um limitador da quantidade de regras que serão feitas. Ao limitar essa quantidade de regras que serão feitas, nós teremos uma diminuição da quantidade de nós puros.   

Então, vamos reaplicar o nosso modelo, dessa vez, executando uma poda na nossa árvore de decisão. Iremos utilizar novamente, na próxima linha: 

``
dt = DecisionTreeClassifier(max_depth=5)
``

Iremos dessa vez passar um parâmetro **max_depth=5** que significa profundidade máxima. E, aqui, podemos dizer a altura máxima da árvore, definido no número 5. Se caso, tivemos uma árvore pequena demais, podemos ter um **underfitting**. Ou se essa árvore tiver nós demais ou uma altura muito grande podemos ter overfitting. Então, cabe muito do nosso cuidado e testes para verifica qual é o tamanho ideal da árvore que iremos utilizar.

O parâmetro max_depth indica a altura máxima da árvore de decisão. Essa altura determina a quantidade de divisões feitas sobre o dado de treino. Diminuindo o número de divisões, é possível evitar uma divisão perfeita dos dados de treino, evitando assim o overfit. Já n_neighbors é um parâmetro do modelo KNN, train_size é um parâmetro da função train_test_split e num_samples não é um parâmetro válido para árvores de decisão.

Ao executarmos esta altura máxima de 5, podemos aplicar novamente o nosso conjunto de dados:

``
X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y) # reaplicando a função treino e teste
``

Assim podemos reconstruir nossa árvore de decisão:

``
dt.fit(X_train,y_train)
``

Na sequência aplicar a função score para verificar os dados de teste.

``
dt.score(X_test,y_test)
``

Então, se mudamos o valor do nosso parâmetro para 3, reaplicar nosso modelo, verificaremos um aumento de acerto na função score. Logo, vale apena continuar executando esses testes de profundidade. 

Terminamos essa aula utilizando o modelo de árvore de decisão e aplicando a técnica de poda, para evitar overfitting. Noutra aula iremos aprender outra perculiaridade da nossa árvore de decisão.

## Selecionando FEATURES com Árvores de Decisão

Nessa aula iremos, aprender a Seleção de Features com Árvores de Decisão. Para isso, iremos utilizar um novo dataset, ou seja, um conjunto de dados com mais colunas e mais atributos, um conjunto de dados de detecção de câncer de mama.

``
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
``

Agora iremos implementar nossa árvore de decisão:

``
dt = DecisionTreeClassifier()
``

Em seguida, iremos separar nosso conjunto de dados, entre atributos e rótulo.

``
X,y = load_breast_cancer(return_X_y=True)
``
Nesse exemplo, não iremos fazer divisão de treino e teste.
Depois, aplicaremos o modelo:

``
dt.fit(X,y)
``

Na próxima linha iremos aplicar o seguinte comando:

``
X.shape
``

Para verificar nosso conjunto de dados que foi carregado.

Então, podemos utilizar o nosso conjunto de dados aplicar, a árvore de decisão nele e verificar as importâncias das **features**.

``
dt.feature_importances_
``

Assim, veremos a importância das features, logo, podemos eliminar algumas features, de menor importância. Para poder adequar melhor nosso conjunto de dados para implementar nosso modelo. E, aí sim, iremos executar isso recursivamente, até possuir um conjunto bem menor de _features_. Essa é uma possibilidade. 

Por que executamos recursivamente, ao invés de retirar todos de uma vez? Porque as _features_ em si podem ter alguma correlação entre elas. Então, retirar uma _feature_ pode mexer na importância de outra. Por isso, vamos retirando uma a uma, recursivamente, até que possamos ter um resultado bom. 

Ou seja,  para eliminar atributos com segurança em uma árvore de decisão, deve-se eliminar o atributo com menor importância, treinar o modelo novamente para verificar como foi a evolução da avaliação do modelo e repetir este processo até que a avaliação do modelo estabilize ou piore. Um especialista pode ajudar na seleção, mas ele estará se utilizando apenas de seu conhecimento e não de uma técnica proveniente da estatística de importância dos atributos. Vale dizer que, eliminar vários atributos de uma vez pode implicar em perdas do modelo, uma vez que atributos de baixa importância podem se correlacionar com outros de alta importância. Por fim, o max_depth evita overfit, não servindo pra eliminação de features.

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.datasets import load_boston, load_iris
from sklearn.model_selection import train_test_split
```

### Árvore de Decisão para  Classificação

```python
dt = DecisionTreeClassifier()
```

```python
X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

### Árvore de Decisão para Regressão

```python
dt = DecisionTreeRegressor()
```

```python
X,y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

## Árvore de decisão com parâmetro max_depth

```python
dt = DecisionTreeClassifier(max_depth=3)
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```
