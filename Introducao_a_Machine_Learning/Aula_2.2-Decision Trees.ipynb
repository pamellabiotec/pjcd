## Implementando a rvore de Deciso

Nessa aula iremos implementar a **rvore de Deciso**, onde iremos aplicar a verso de classsificao quanto a verso de regresso.

Ento, iremos iniciar importando as bibliotecas que iremos utilizar.


``
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
``

Dessa maneira, estamos importando a biblioteca de rvore de deciso tanto para classificao quanto para regresso na mesma linha. Em seguida, iremos importar os datasets que iremos utilizar. Sendo Boston para regresso e Iris para classificao.


``
from sklearn.datasets import load_boston, load_iris
``

Na sequncia iremos aplicar a diviso de treino e teste.


``
from sklearn.model_selection import train_test_split
``

Para implementar a rvore de deciso para classificao utilizaremos o seguinte comando:


``
dt = DecisionTreeClassifier()
``

Em seguida, iremos dividir os conjunto de dados Iris em treino e teste usando a funo train_test_split.

``
X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y)
``

Onde X receber os atributos e o y receber o rtulo dos dados. Em seguida, iremos aplicar a funo de diviso de treino e teste.

Na sequncia, implementaremos a rvore de deciso para classificao:

``
dt.fit(X_train,y_train)
``

Ao executar essa linha, foi treinada nossa rvore de deciso para classificao.  Agora, iremos utilizar a funo **score** para verificar essa implementao.

``
dt.score(X_test,y_test)
``

Ao aplicar essa funo, teremos como resultado 94% de preciso. Agora, iremos aplicar a rvore de deciso para regresso utilizando passos bem parecidos. 


``
dt = DecisionTreeRegressor()
``

Assim, instanciamos nosso modelo para regresso.
Nesse momento, iremos carregar o dataset Boston com seguinte cdigo:

``
X,y = load_boston(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X,y)

``

Na sequncia implementaremos a rvore de deciso para regresso:


``
dt.fit(X_train,y_train)
``

E verificaremos nosso modelo com a funo score:


``
dt.score(X_test,y_test)
``

Assim, teremos como resultado que nosso modelo teve 69% de acerto.

O primeiro ponto que precisamos verificar  que a rvore de deciso por natureza j possui um alto **ndice de overfitting**, porque , pela sua natureza ela tenta sempre carregar todo o conjunto de dados etenta separ-lo perfeitamente. Ento, teremos que aplicar alguma tcnica para minizar o efeito de overfitting. 

Logo, a partir deste momento, irei retreinar a rvore de deciso utilizando um limitador da quantidade de regras que sero feitas. Ao limitar essa quantidade de regras que sero feitas, ns teremos uma diminuio da quantidade de ns puros.   

Ento, vamos reaplicar o nosso modelo, dessa vez, executando uma poda na nossa rvore de deciso. Iremos utilizar novamente, na prxima linha: 

``
dt = DecisionTreeClassifier(max_depth=5)
``

Iremos dessa vez passar um parmetro **max_depth=5** que significa profundidade mxima. E, aqui, podemos dizer a altura mxima da rvore, definido no nmero 5. Se caso, tivemos uma rvore pequena demais, podemos ter um **underfitting**. Ou se essa rvore tiver ns demais ou uma altura muito grande podemos ter overfitting. Ento, cabe muito do nosso cuidado e testes para verifica qual  o tamanho ideal da rvore que iremos utilizar.

O parmetro max_depth indica a altura mxima da rvore de deciso. Essa altura determina a quantidade de divises feitas sobre o dado de treino. Diminuindo o nmero de divises,  possvel evitar uma diviso perfeita dos dados de treino, evitando assim o overfit. J n_neighbors  um parmetro do modelo KNN, train_size  um parmetro da funo train_test_split e num_samples no  um parmetro vlido para rvores de deciso.

Ao executarmos esta altura mxima de 5, podemos aplicar novamente o nosso conjunto de dados:

``
X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y) # reaplicando a funo treino e teste
``

Assim podemos reconstruir nossa rvore de deciso:

``
dt.fit(X_train,y_train)
``

Na sequncia aplicar a funo score para verificar os dados de teste.

``
dt.score(X_test,y_test)
``

Ento, se mudamos o valor do nosso parmetro para 3, reaplicar nosso modelo, verificaremos um aumento de acerto na funo score. Logo, vale apena continuar executando esses testes de profundidade. 

Terminamos essa aula utilizando o modelo de rvore de deciso e aplicando a tcnica de poda, para evitar overfitting. Noutra aula iremos aprender outra perculiaridade da nossa rvore de deciso.

## Selecionando FEATURES com rvores de Deciso

Nessa aula iremos, aprender a Seleo de Features com rvores de Deciso. Para isso, iremos utilizar um novo dataset, ou seja, um conjunto de dados com mais colunas e mais atributos, um conjunto de dados de deteco de cncer de mama.

``
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
``

Agora iremos implementar nossa rvore de deciso:

``
dt = DecisionTreeClassifier()
``

Em seguida, iremos separar nosso conjunto de dados, entre atributos e rtulo.

``
X,y = load_breast_cancer(return_X_y=True)
``
Nesse exemplo, no iremos fazer diviso de treino e teste.
Depois, aplicaremos o modelo:

``
dt.fit(X,y)
``

Na prxima linha iremos aplicar o seguinte comando:

``
X.shape
``

Para verificar nosso conjunto de dados que foi carregado.

Ento, podemos utilizar o nosso conjunto de dados aplicar, a rvore de deciso nele e verificar as importncias das **features**.

``
dt.feature_importances_
``

Assim, veremos a importncia das features, logo, podemos eliminar algumas features, de menor importncia. Para poder adequar melhor nosso conjunto de dados para implementar nosso modelo. E, a sim, iremos executar isso recursivamente, at possuir um conjunto bem menor de _features_. Essa  uma possibilidade. 

Por que executamos recursivamente, ao invs de retirar todos de uma vez? Porque as _features_ em si podem ter alguma correlao entre elas. Ento, retirar uma _feature_ pode mexer na importncia de outra. Por isso, vamos retirando uma a uma, recursivamente, at que possamos ter um resultado bom. 

Ou seja,  para eliminar atributos com segurana em uma rvore de deciso, deve-se eliminar o atributo com menor importncia, treinar o modelo novamente para verificar como foi a evoluo da avaliao do modelo e repetir este processo at que a avaliao do modelo estabilize ou piore. Um especialista pode ajudar na seleo, mas ele estar se utilizando apenas de seu conhecimento e no de uma tcnica proveniente da estatstica de importncia dos atributos. Vale dizer que, eliminar vrios atributos de uma vez pode implicar em perdas do modelo, uma vez que atributos de baixa importncia podem se correlacionar com outros de alta importncia. Por fim, o max_depth evita overfit, no servindo pra eliminao de features.

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.datasets import load_boston, load_iris
from sklearn.model_selection import train_test_split
```

### rvore de Deciso para  Classificao

```python
dt = DecisionTreeClassifier()
```

```python
X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

### rvore de Deciso para Regresso

```python
dt = DecisionTreeRegressor()
```

```python
X,y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

## rvore de deciso com parmetro max_depth

```python
dt = DecisionTreeClassifier(max_depth=3)
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```
